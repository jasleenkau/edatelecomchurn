**Pie chart** - A pie chart in Python is a circular statistical graphic that is divided into slices to illustrate numerical proportion. In a pie chart, the arc length of each slice is proportional to the quantity it represents.

```python
churn_counts = telecom['Churn'].value_counts()
fig1, ax1 = plt.subplots()
ax1.pie(churn_counts, labels = ['False', 'True'], autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Telecom Churn Data Analysis')
plt.show()




```

**Count plot**- , A count plot chart is a type of categorical plot used to show the count of observations in each category of a categorical variable. It is typically created using the Seaborn library, which is built on top of Matplotlib and provides a high-level interface for creating attractive and informative statistical graphics.

The count plot is essentially a bar plot where the x-axis represents the categories of the categorical variable, and the height of each bar represents the count of occurrences of each category in the dataset. It is especially useful for visualizing the distribution and frequency of categorical data.

```python
sns.countplot(x='Churn', data=telecom) # The countplot is used to represent the occurrence(counts) of churn present.
plt.title('Customer Churn Distribution') #It uses the concept of a bar chart for the visual depiction.
plt.show()
telecom['Churn'].value_counts()
```
